```python
# Get all dates for which you want to pull players ranking, change periods value for that
import pandas as pd
from datetime import datetime

dates = pd.date_range(end = datetime.today(), periods = 20).to_pydatetime().tolist()
dates = [d.strftime('%Y-%m-%d') for d in dates]
print(dates)
```

```python
# prepare request url where date will change in request
odi_batsman_ranking_request = 'https://cricketapi-icc.pulselive.com/icc-ratings/ranked/players/odi/bat?pageSize=10&dmsPlayerIdRequired=true&at=%s'
odi_bowler_ranking_request = 'https://cricketapi-icc.pulselive.com/icc-ratings/ranked/players/odi/bowl?pageSize=10&dmsPlayerIdRequired=true&at=%s'
odi_all_rounder_ranking_request = 'https://cricketapi-icc.pulselive.com/icc-ratings/ranked/players/odi/allround?pageSize=10&dmsPlayerIdRequired=true&at=%s'
```

```python
# iterate through dates, send the request and store the ranking response
# using dict to store the ranking table, where date is key and value is ranking list
# ranking list in dict is sorted as per rank
odi_batsman_ranking_dict = {}
odi_bowler_ranking_dict = {}
odi_all_rounder_ranking_dict = {}

import requests
import json

for day in dates:
    odi_bat_request_url = odi_batsman_ranking_request%day
    # print(odi_bat_request_url)
    odi_bat_response = requests.get(odi_bat_request_url).text
    odi_bat_response = json.loads(str(odi_bat_response))
    odi_batsman_ranking_dict[day] = [player['player'] for player in odi_bat_response['content']]
    
    odi_bowler_request_url = odi_bowler_ranking_request%day
    # print(odi_bowler_request_url)
    odi_bowler_response = requests.get(odi_bowler_request_url).text
    odi_bowler_response = json.loads(str(odi_bowler_response))
    odi_bowler_ranking_dict[day] = [player['player'] for player in odi_bowler_response['content']]
    
    odi_all_rounder_request_url = odi_all_rounder_ranking_request%day
    # print(odi_all_rounder_request_url)
    odi_all_rounder_response = requests.get(odi_all_rounder_request_url).text
    odi_all_rounder_response = json.loads(str(odi_all_rounder_response))
    odi_all_rounder_ranking_dict[day] = [player['player'] for player in odi_all_rounder_response['content']]
```
